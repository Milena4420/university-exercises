## Exercice 1
Le problème de la ligne d'horizon. Un immeuble Bi est représenté par un triplet (Li, Hi, Ri) où Li et Ri désignent la coordonées x de la partie gauche et droite de l'immeuble, et Hi désigne la hauteur de l'immeuble. Une ligne d'horizon d'un ensemble de n immeuble est une liste des coordonnées x et des hauteurs qui les relient, classées dans l'ordre de gauche à droite.

Exemple: La ligne d'horizon des immeubles

<center>{(3, 13, 9),(1, 11, 5),(12, 7, 16),(14, 3, 25),(19, 18, 22),(2, 6, 7),(23, 13, 29),(23, 4, 28)}</center>

<center>est</center>

<center>{1, 11, 3, 13, 9, 0, 12, 7, 16, 3, 19, 18, 22, 3, 23, 13, 29, 0}</center>

<center>où les points référencent alternativement l'axe des abscissses et des ordonnées</center>

<center>(notez que les coordonnées x dans un horizon sont triées).</center>

![skylines.png](./skylines.png)

Soit la taille d'une ligne d'horizon le nombre total d'éléments (coordonnées et hauteurs) de sa liste. Implémentez un algorithme pour combiner une ligne d'horizon A de taille n1 et une ligne d'horizon B de taille n2 en une ligne d'horizon S de taille O(n1 + n2). Votre algorithme doit s'exécuter en un temps O(n1 + n2).

```python
def combine_skyline (a,b):
    i,j=0,0 #on initialise nos indexes de a et b à 0
    ha, hb = 0,0 #on initialise nos hauteurs courantes de à et b à 0
    lasth=0 #on initialise la dernière hauteur donnée
    horizonS = [] #on crée une liste vide qui sera notre résultat

    while (i<len(a) or j<len(b)):
        if j>=len(b) or (i<len(a) and a[i]<b[j]): 
            x = a[i]
            ha = a[i+1] #on change la hauteur courante de a
            i +=2 #on incrémente de 2 puisqu'on regarde les éléments par paire

        elif i>=len(a) or (j<len(b) and b[j]<a[i]):
            x = b[j]
            hb = b[j+1] #on change la hauteur courante de b
            j +=2 #on incrémente de 2 puisqu'on regarde les éléments par paire

        else:
            x = a[i]
            ha = a[i+1] #on change la hauteur courante de a
            hb = b[j+1] #on change la hauteur courante de b
            i +=2 #on incrémente de 2 puisqu'on regarde les éléments par paire
            j +=2 #on incrémente de 2 puisqu'on regarde les éléments par paire

        h = max(ha, hb)
        
        if h!= lasth:
            if horizonS and horizonS[-2]==x:
                horizonS[-1] = h
            else:
                horizonS.extend([x,h])
            lasth = h
    return horizonS
          
```

```python
a = [1,8,6,10,8,0]
b = [2,4,5,15,11,0]
s = combine_skyline(a,b)
# Décommentez la ligne ci-dessous si votre réponse est une liste en 2D pour la convertir en liste 1D
# s = sum(s, [])
print("Skyline A:")
print(a)
print("Skyline B:")
print(b)
print("Skyline S (must be [1, 8, 5, 15, 11, 0]):")
print(s)
assert s == [1,8,5,15,11,0], "The merge of skyline A and skyline B failed"

print()

a = [2,7,7,9,10,5,14,12,15,0]
b = [3,6,8,12,9,3,13,15,16,0]
s = combine_skyline(a,b)
# Décommentez la ligne ci-dessous si votre réponse est une liste en 2D pour la convertir en liste 1D
# s = sum(s, [])
print("Skyline A:")
print(a)
print("Skyline B:")
print(b)
print("Skyline S (must be [2, 7, 7, 9, 8, 12, 9, 9, 10, 5, 13, 15, 16, 0]):")
print(s)
assert s == [2,7,7,9,8,12,9,9,10,5,13,15,16,0]
```

### Explications

- Le but est d'entrer (et rester) dans la boucle while tant qu'on a pas visité tout a et tout b. 
- 1e if: on y entre lorsqu'il n'y a plus que a à visiter ou que la position de l'élément de a est plus petite que celle b. Dans ce cas, on regarde la position et la hauteur dans a.
- 2e if: on y entre lorsqu'il n'y a plus que b à visiter ou que la position de l'élément de b est plus petite que celle a. Dans ce cas, on regarde la position et la hauteur dans b.
- else: c'est le cas ou la position de a et b changent au même x, on prend en compte les 2 hauteurs.
- on pose h comme le maximum entre les deux dernière hauteurs de a et b. Le but du dernier if est de regarder si h est différente que la dernière hauteur enregistrée. Si la position actuelle est la même que la derniere position enregistrée dans la horizonS, on change la valeur de sa hauteur. Sinon, on ajoute la paire (position, hauteur) à horizonS.
- L'algorithme est en O(n1+n2) car dans chaque itération on avance i ou j (ou les deux) donc il y a au plus n1+n2 itérations. Toutes les autres opérations (comparaison, ajout dans horizonS,...) se font en temps constant.

### Exercice 1.1
Implémentez un algorithme de "diviser pour régner" O(n log n) pour trouver la ligne d'horizon de n bâtiments.


```python
def find_skyline(buildings):
    
    if len(buildings)==0:  #le cas où il n'y a pas de bâtiment, on met ce cas en évidence sinon il fera une récursivité infinie 
        return []
    
    if len(buildings)==1:
        L, H, R = buildings[0]
        return [L, H, R, 0] #on met la hauteur à 0 à la fin du bâtiment
    
    mid = len(buildings)//2 #on pose un milieu pour pouvoir diviser le problème en 2 parties: côté gauche et côté droite
    left = find_skyline(buildings[:mid])
    right = find_skyline(buildings[mid:])

    return combine_skyline(left, right) #on utilise la fonction précédente pour combiner les 2 skylines (left, right) qu'on aura trouvé 
```

```python
skyline = find_skyline([[3, 13, 9],[1, 11, 5],[12, 7, 16],[14, 3, 25],[19, 18, 22],[2, 6, 7],[23, 13, 29],[23, 4, 28]])
# Décommentez la ligne ci-dessous si votre réponse est une liste en 2D pour la convertir en liste 1D
# skyline = sum(skyline, [])
print (skyline)
assert skyline == [1, 11, 3, 13, 9, 0, 12, 7, 16, 3, 19, 18, 22, 3, 23, 13, 29, 0]
```

## Exercice 2
Le tri par fusion (merge sort) est l'algorithme classique de type "diviser pour régner". Implémentez et analysez-le en essayant de grands inputs. Assurez-vous d'essayer des inputs à la fois distribués de manière aléatoire et triées par ordre décroissant.

```python
def merge_sort(s):
    
    if len(s) <= 1:
        return s

    mid = len(s) // 2
    left = merge_sort(s[:mid])
    right = merge_sort(s[mid:])

    i, j = 0, 0
    merged = []

    while i < len(left) and j < len(right):
        if left[i] <= right[j]:
            merged.append(left[i])
            i += 1
        else:
            merged.append(right[j])
            j += 1

    merged.extend(left[i:])
    merged.extend(right[j:])

    return merged

import random
random_list = []
for i in range(0,20):
    n = random.randint(1,5000)
    random_list.append(n)

print("Initial list (random order): ")
print(random_list)
print("My sorted list")
my_sorted_list = merge_sort(random_list.copy())
print(my_sorted_list)
print("Python sorted list")
python_sorted_list = random_list.copy()
python_sorted_list.sort()
print(python_sorted_list)
assert my_sorted_list == python_sorted_list, "The list is not correctly sorted"

print()
random_list = []
for i in range(0,25):
    n = random.randint(1,5000)
    random_list.append(n)

print("Initial list (descending order): ")
random_list.sort()
random_list.reverse()
print(random_list)
print("My sorted list")
my_sorted_list = merge_sort(random_list.copy())
print(my_sorted_list)
print("Python sorted list")
python_sorted_list = random_list.copy()
python_sorted_list.sort()
print(python_sorted_list)
assert my_sorted_list == python_sorted_list, "The list is not correctly sorted"

```

### Explications

```python
"""
    La fonction merge_sort reçoit en paramètre s : une liste d'entiers. 
    Elle renvoie merged : une nouvelle liste contenant les éléments de s classés dans l'ordre croissant.

    L’idée du code est d’exploiter le fait qu’il est très efficace de fusionner deux listes déjà triées. 
    Donc, pour une liste donnée, on va la séparer récursivement en deux au milieu pour obtenir deux sous-listes disjointes, 
    de même taille à un élément près. Puis on trie chaque moitié dans l'ordre croissant. 
    Finalement, il suffit de fusionner ces deux sous-listes.

    Nous pouvons maintenant analyser le code.
    Notre fonction merge_sort reçoit une liste d'entiers.
    Puisque nous voulons créer une fonction récursive, nous devons définir un cas de base. 
    Si la liste contient au plus un élément, alors nous n'avons pas besoin de la trier : nous pouvons simplement la renvoyer.

    Ensuite, nous définissons mid, l'indice du milieu. 
    Étant donné que les indices d'une liste contenant au moins deux éléments prennent des valeurs dans {0, 1, ..., len(s) - 1}, 
    avec len(s) le nombre d'éléments de s, il suffit de prendre la partie entière de len(s) divisé par deux.

    Par récursion, nous allons récupérer la première moitié de la liste triée dans l'ordre croissant en appelant notre fonction 
    sur cette sous-liste.  Puis nous récupérons l'autre moitié triée en appelant la fonction sur le reste de la liste. Si le nombre d'éléments de 
    la liste est impair, l’élément du milieu est inclus dans la seconde moitié (s[mid:]).

    Il ne reste plus qu'à fusionner ces deux moitiés triées.
    Nous allons construire i et j, qui vont respectivement agir comme des pointeurs : 
    l’un pour la première moitié de la liste (left), l’autre pour la seconde (right).
    Ils représentent les indices des éléments actuellement considérés dans chaque sous-liste. 
    Au début, ils pointent tous les deux sur le premier élément de leur liste respective (indice 0).

    Nous créons ensuite merged, une liste vide, qui va devenir la liste triée finale.

    Nous avons ensuite la boucle principale. 
    Tant que les deux listes contiennent encore des éléments à traiter (c’est-à-dire tant que i < len(left) et j < len(right)), 
    si l'élément de la première moitié sur lequel pointe i est plus petit ou égal à l'élément de la seconde moitié sur lequel pointe j, 
    alors nous ajoutons cet élément de left à merged, puis nous décalons le pointeur i en l'incrémentant de 1. 
    Sinon, nous ajoutons l'élément correspondant de right à merged et nous incrémentons j de 1.

    Une fois qu'une des deux listes a été entièrement traitée, nous sortons de la boucle. 
    Il suffit alors d'ajouter à merged tous les éléments restants de l'autre liste (left ou right), 
    car ils sont déjà triés et sont tous plus grands ou égaux aux éléments déjà ajoutés. 
    Enfin, nous retournons merged.

"""
```

```python
import random
import time

def merge_sort(s):
    if len(s) <= 1:
        return s

    mid = len(s) // 2
    left = merge_sort(s[:mid])
    right = merge_sort(s[mid:])

    i, j = 0, 0
    merged = []

    while i < len(left) and j < len(right):
        if left[i] <= right[j]:
            merged.append(left[i])
            i += 1
        else:
            merged.append(right[j])
            j += 1

    merged.extend(left[i:])
    merged.extend(right[j:])

    return merged

for k in range(10):
    random_list = [random.randint(1, 10**6) for _ in range(50000)]
    start = time.time()
    merge_sort(random_list)
    end = time.time()
    print("Temps pour liste aléatoire :", end - start, "secondes")
    
    
    descending_list = sorted(random_list, reverse=True)
    start = time.time()
    merge_sort(descending_list)
    end = time.time()
    print("Temps pour liste décroissante :", end - start, "secondes")
    print()
    print()
    print()
```
